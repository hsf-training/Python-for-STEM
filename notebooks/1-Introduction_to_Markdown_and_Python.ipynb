## Links
Sometimes linking to another website is necesarry in situations where one needs to make a reference or point to a source of information. This is especially useful for tutorials and assignments.

to make a link in Markdown use the following syntax

~~~
[text to click on](hhtps://someurl.com)
~~~

[Link to a Great Markdown Reference](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet#links)


Here are all the methods one can create links

```
[I'm an inline-style link](https://www.google.com)

[I'm an inline-style link with title](https://www.google.com "Google's Homepage")

[I'm a reference-style link][Arbitrary case-insensitive reference text]


[You can use numbers for reference-style link definitions][1]

Or leave it empty and use the [link text itself].

URLs and URLs in angle brackets will automatically get turned into links. 
http://www.example.com or <http://www.example.com> and sometimes 
example.com (but not on Github, for example).

Some text to show that the reference links can follow later.

[arbitrary case-insensitive reference text]: https://www.mozilla.org
[1]: http://slashdot.org
[link text itself]: http://www.reddit.com
```

### Basics #3 - Numeric Operations

````
adding = x + y

subtract = x - y

division = x / y

modulus = x % y

print ("x + y = {}".format(adding))
print ("x - y = {}".format(subtract))
print ("x / y = {}".format(division))
print('x % y = {}'.format(modulus))
````

### Basics #6 - Collections

````
# Tuple
tup = (1, 2, 3, 4)

# List
lst = [1, 2, 3, 4]
````

````
# Tuples are IMMUTABLE
# They cannot be re-assigned once declared!

# tup[1] = 10 will return an error on execution!!
````

````
# Lists are MUTABLE
# They can be altered

lst[1] = 10

print ("List after element at index 1 was reassigned:")
print (lst)

print ("Sliced list from index=1 -> index=3:")
print (lst[1:3])

lst.pop()
print ("List after executing pop():")
print (lst)


lst.remove(10)
print ('List after removing elemtent "10":')
print (lst)
````

````
# Strings are treated as a Tuple (IMMUTABLE)

stringTest = "Happy Birthday!"

# string substring from index 0..5
print (stringTest[0:5])

# string substring from index 6..n-1
print (stringTest[6:])
````

### Basics #7 - Interating Collections

````
# for loop from 0 to n
print("for loop from 0 to n:")
for x in range(10):
  print (x)

print()
print("for loop using defined range:")
# for loop using defined range
for x in range(2, 5):
  print (x)
````

````
lst = ["dog", "cat", "airplane", "box"]
print ("for loop interating through elements in a collection:")
for x in lst:
  print (x)

if "dog" in lst:
  print ("\n'dog' found in list!")
  
if "student" not in lst:
  print ("'student' not found in list!")
````
